```python
import random
import gymnasium as gym
import pandas as pd

# Environmnet
env_name = "CartPole-v1"
env = gym.make(env_name)
d_states = env.observation_space.shape[0]
n_actions = env.action_space.n
```

```python
# one way sampling
data = []
for i in range(100):
    state, info = env.reset()
    done = False
    episodic_return = 0  # episodic return

    while not done:
        # a = random.randrange(0,2)
        a = 1  # one way policy
        next_state, reward, done = env.step(a)[:3]
        episodic_return += reward

    data.append(episodic_return)

df = pd.DataFrame(data)
df.describe().round(2)
```

```python
# Random_sampling
data = []
for i in range(100):
    state, info = env.reset()
    done = False
    episodic_return = 0  # episodic return

    while not done:
        a = random.randrange(0,2)
        # a = 1  # one way policy
        next_state, reward, done = env.step(a)[:3]
        episodic_return += reward

    data.append(episodic_return)

df = pd.DataFrame(data)
df.describe().round(2)
```
